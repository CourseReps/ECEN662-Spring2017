```python
import math
import numpy as np
import pandas as pd

sample_df = pd.DataFrame.from_csv("Test1/Data1Set3_viswamnathanTAMU.csv")

#d Y_hat[]
#Y_hat = sample_df
Y_hat = pd.DataFrame(np.zeros((len(sample_df.index), 1)))
Y_hat.columns = ['Y']

for index, row in sample_df.iterrows() :
    fY1 = 1
    fY0 = 1
    for index2 in range(0,len(sample_df.columns)) :
        dataValue = sample_df.iloc[index][index2]
        fY1 = fY1*(1/math.sqrt(2*math.pi))*math.exp(-(((dataValue-1)**2)/2))
        fY0 = fY0*(1/math.sqrt(2*math.pi))*math.exp(-(((dataValue)**2)/2))
        #print(fY1)
        #print(fY0)
    Y_hat.iloc[index][0] = (fY1>fY0)    
        
print(Y_hat)
#Y_hat = (sample_df > 0.5)
Y_hat['Y'] = Y_hat['Y'].map({False: 0, True: 1})
print(Y_hat)

Y_hat.to_csv("Answer0.csv")
```
