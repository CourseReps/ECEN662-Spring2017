Problem : To find a hypotheis testing rule to distinguish betweeen Natural and computer generated scenery.

To formulate a binary hypothesis testing rule for the given datasets of images, the Red, Green and Blue Values of individual images are extracted for each image.

Using these extracted values, the absolute value of difference in between these values for the adjacent pixels for each image is calculated.

The average of the absolute differences over all pixels gives us the Average R,G,B values for an image.

Now, by fitting these average values to a Normal distribution gives us the PDF Value for each of the images.The Mean and Variance are previously computed from the R,G,B values for each of the Datasets.

The Hypothesis testing rule then turns out to be taking the likelihood ratio test, i.e., 
    
LRT : Natural/Synthetic ratio  >||< = 1.

Based on the Outcome of the likelihood ratio test, each image is classified as either Natural or Synthetic.

The Error ratio in Classification, is then computed as (Total Number of Errors)/(Total Number of images classified).
Approximately 1.17% of images are classified as errors for the given datasets.

# CODE:

Initialization

```python
import os
from PIL import Image
import numpy as np
from scipy.stats import norm
import math
```

The following code initializes the Average Mean and Variances of R, G, B Values for Natural & Synthetic Datasets. These values are generated using the code in "RGB_Generator" File.

```python
# For Natural Dataset

#Real, r - g - b
#
#50.998743651795749 , 246.26951426278168
#48.854255260942764 , 319.06770996344602
#48.414790333894501 , 409.7224373340897

rrMean = 50.998743651795749
rrVar = math.sqrt(246.26951426278168)

rgMean = 48.854255260942764
rgVar = math.sqrt(319.06770996344602)

rbMean = 48.414790333894501
rbVar = math.sqrt(409.7224373340897)

# For Synthetic Dataset

#Synthetic, r - g - b
#91.481063503086432 , 103.73098113899538
#92.321147569444491 , 105.994811365883
#91.978386332947537 , 88.778210322054306

srMean = 91.481063503086432
srVar =math.sqrt(103.73098113899538)

sgMean = 92.321147569444491 
sgVar = math.sqrt(105.994811365883)

sbMean = 91.978386332947537
sbVar = math.sqrt(88.778210322054306)


errors = 0.0
total = 0.0
```

Running the Hypothesis test on "Training Set Scenes Data" and computing errors

```python
for filename in os.listdir('TrainingSetScenes'):
    print(filename)
    img = Image.open("TrainingSetScenes/" + filename).convert('RGB').resize((960,540), Image.ANTIALIAS)
    im = np.reshape(np.array(img.getdata()), [960, 540, 3])
    rval = 0.0
    gval = 0.0
    bval = 0.0
    for i in range(0,540):
        for j in range(0,960-1):
            rval += abs(im[j+1, i, 0] - im[j,i,0])
            gval += abs(im[j+1, i, 1] - im[j,i,1])
            bval += abs(im[j+1, i, 2] - im[j,i,2])
    rval *= (1.0/(540*960))
    gval *= (1.0/(540*960))
    bval *= (1.0/(540*960))
    lReal = norm.pdf(rval, rrMean, rrVar)*norm.pdf(gval, rgMean, rgVar)*norm.pdf(bval, rbMean, rbVar)
    print(lReal)
    lSynth = norm.pdf(rval, srMean, srVar)*norm.pdf(gval, sgMean, sgVar)*norm.pdf(bval, sbMean, sbVar)
    print(lSynth)
    if lSynth > lReal:
    	errors += 1.0
    	total += 1.0
    else:
    	total += 1.0
```

Running the Hypothesis test on "Training Set Synthetic Data" and computing errors

```python
for filename in os.listdir('TrainingSetSynthetic'):
    print(filename)
    img = Image.open("TrainingSetSynthetic/" + filename).convert('RGB').resize((960,540), Image.ANTIALIAS)
    im = np.reshape(np.array(img.getdata()), [960, 540, 3])
    rval = 0.0
    gval = 0.0
    bval = 0.0
    for i in range(0,540):
        for j in range(0,960-1):
            rval += abs(im[j+1, i, 0] - im[j,i,0])
            gval += abs(im[j+1, i, 1] - im[j,i,1])
            bval += abs(im[j+1, i, 2] - im[j,i,2])
    rval *= (1.0/(540*960))
    gval *= (1.0/(540*960))
    bval *= (1.0/(540*960))
    #print(rval)
    lReal = norm.pdf(rval, rrMean, rrVar)*norm.pdf(gval, rgMean, rgVar)*norm.pdf(bval, rbMean, rbVar)
    print(lReal)
    lSynth = norm.pdf(rval, srMean, srVar)*norm.pdf(gval, sgMean, sgVar)*norm.pdf(bval, sbMean, sbVar)
    print(lSynth)
    if lSynth > lReal:
    	total += 1.0
    else:
    	errors += 1.0
    	total += 1.0
```

Error Calculation:

```python
# Total Number of Images (Scenery + Synthetic)
print(total)
```

```python
# Total Number of Errors
print(errors)
```

Error Ratio:

```python
# Error Ratio
print(errors/total)
```

```python
# Error Percentage
print(errors*100/total)
```
