```python
print("Helloworld");
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.mlab as mlab
import math
from scipy.stats import norm


fig, ax1 =plt.subplots()

m=np.linspace(2,6,4)
for l in range(4):
  mu0 = -1
  variance0 = 1
  sigma0 = math.sqrt(variance0)
  x = np.linspace(-4, 4, 200)
  ax1.plot(x, mlab.normpdf(x, mu0, sigma0))
  mu1= m[l]
  variance1=3
  sigma1=math.sqrt(variance1)
  ax1.plot(x,mlab.normpdf(x,mu1,sigma1),color='r')
  ax2=ax1.twinx()
  log_like=np.log(mlab.normpdf(x,mu1,sigma1))-np.log(mlab.normpdf(x,mu0,sigma0));
  tou=np.linspace(-2.5,11,100)
  prob_false_alarm=[0]*100
  prob_det=[0]*100
  for k in range(100):
    #print(k)
    y = [0]*200
    for i in range(200):
      y[i] = -1.5
    inter=log_like-tou[k];
    #print(inter)
    point=[]
    j=2
    while j<200:
     if ((inter[j]>0 and inter[j-1]<0) or (inter[j]<0 and inter[j-1]>0) ):
        point.append(j)
     j=j+1
    #print(point)
    if len(point)==2:
      prob_false_alarm[k]=norm.cdf(x[point[0]],mu0,sigma0)-norm.cdf(x[point[1]],mu0,sigma0)+1
      p=norm.cdf(x[point[1]],mu1,sigma1)-norm.cdf(x[point[0]],mu1,sigma1)
      prob_det[k]=1-p
    else:
      prob_det[k]=1-norm.cdf(x[point[0]],mu1,sigma1)
      prob_false_alarm[k]=1-norm.cdf(x[point[0]],mu0,sigma0)

    #print(prob_false_alarm)
    #print(prob_det)
  ax2.plot(x,np.log(mlab.normpdf(x,mu1,sigma1))-np.log(mlab.normpdf(x,mu0,sigma0)),'k--')
  plt.figure(2)
  ax3=plt.subplot(111)
  ax3.set_title("ROC curves")
  ax3.set_xlabel("Probability of FALSE ALARM")
  ax3.set_ylabel("Probabilty of DETECTION")
  if l==0:
    ax3.plot(prob_false_alarm,prob_det,label='mean_red=2, mean_blue=-1')
  elif l==1:
    ax3.plot(prob_false_alarm,prob_det,label='mean_red=3.33, mean_blue=-1')
  elif l==2:
    ax3.plot(prob_false_alarm,prob_det,label='mean_red=4.6, mean_blue=-1')
  else :
    ax3.plot(prob_false_alarm,prob_det,label='mean_red=6, mean_blue=-1')
  ax3.text(0.2,0.6,'mean_red=2,3.3,4.6,6 mean_blue=-1')
        
plt.show()
```
