```python
import pandas as pd
import numpy as np
import matplotlib.mlab as mlab
practice_data=pd.DataFrame.from_csv("Data1Set0_richarddoss.csv")
#print(practice_data)
Ze_Hat=((mlab.normpdf(practice_data,1,1)/(mlab.normpdf(practice_data,0,1)))>1)
print(Ze_Hat)
```

```python
Ze_Hat['Y'] = Ze_Hat['Y'].map({False: 0, True: 1})
correct_ans=pd.DataFrame.from_csv("Data1Solution0_richarddoss.csv")
print(Ze_Hat['Y'])
print(correct_ans['Z'])
print(np.count_nonzero(correct_ans['Z']==Ze_Hat['Y']))
Ze_Hat.to_csv("Practice0_solutions.csv")
```
