```python
# Likelihood Function Plot
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.mlab as mlab
import math

# plot laplace distributions
fig, ax1 = plt.subplots()
ax1.set_ylabel('Laplacian Distributions')
ax1.set_title('Likelihood Ratio (Laplace Distributions)')
mean, variance = -1., 2.
scale = np.sqrt(variance/2)

mean2, variance2 = 1., 2.
scale2 = np.sqrt(variance2/2)

x = np.arange(-8., 8., .01)
pdf = np.exp(-abs(x-mean)/scale)/(2.*scale)
ax1.plot(x, pdf)

pdf2 = np.exp(-abs(x-mean2)/scale2)/(2.*scale2)
ax1.plot(x, pdf2)

# plot likelihood ratio curve (positive mean over negative mean)

ax2 = ax1.twinx()
ax2.set_ylabel('Likelihood Ratio (Logscaled)')
lhood_ratio = np.log(pdf2)-np.log(pdf)
ax2.plot(x,lhood_ratio,'k--')

plt.show()

```

```python
# ROC Curve Plot
```

```python
fig2, ax3 = plt.subplots()

false_pos = []
true_pos = []

for value in x:
    if value >= mean2:
        true_pos.append(1/2*np.exp(-(value-mean2)/scale2))
    else:
        true_pos.append(1-1/2*np.exp((value-mean2)/scale2))
        
    if value >= mean:
        false_pos.append(1/2*np.exp(-(value-mean)/scale))
    else:
        false_pos.append(1-1/2*np.exp((value-mean)/scale))

ax3.plot(false_pos,true_pos)
ax3.set_title('ROC Curve (Laplace Distributions)')
ax3.set_xlabel('False Positive')
ax3.set_ylabel('True Positive')

plt.show()
```
