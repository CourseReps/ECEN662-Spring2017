# Probability False Alarm /Detection

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.mlab as mlab
import math
from scipy.stats import laplace
import scipy.special
import numpy as np
qfunc = lambda x: 0.5-0.5*scipy.special.erf(x/np.sqrt(2))


from scipy.stats import norm

fig, ax1 = plt.subplots()

x = np.linspace(-100, 100, 2000)
mu = np.linspace(0.25, 1.5, 6)
for index in range(6):
    mu0 = -mu[index]


    pfa=qfunc((x/2*mu0)+mu0)
    pd=qfunc((x/2*mu0)-mu0)

    ax1.plot(pd,pfa)

plt.xlabel('P False Alarm')
plt.ylabel('P Detection')
plt.title('Probability of detection vs probability of false alarm')
plt.show()
```
