```python
# Initialisation
# Importing required functions
import math
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
from scipy.stats import norm, laplace, cauchy
```

# Example ROC curves for Laplace

```python
variance0 = 1
sigma0 = math.sqrt(variance0)
variance1 = 1
sigma1 = math.sqrt(variance1)


fig, ax1 = plt.subplots()

x = np.linspace(-4, 4, 200)
mu = np.linspace(0.5, 2.0, 6)
for index in range(5):
    mu0 = -mu[index]
    mu1 = mu[index]
    ax1.plot(laplace.cdf((x-mu1)/sigma1),laplace.cdf((x-mu0)/sigma0))
    
plt.title('ROC Curves for Laplace Distribution')
plt.show()
```

# Example ROC curves for Cauchy

```python
variance2 = 1
sigma2 = math.sqrt(variance2)
variance3 = 1
sigma3 = math.sqrt(variance3)


fig, ax1 = plt.subplots()

x = np.linspace(-4, 4, 200)
mu = np.linspace(0.5, 2.0, 6)
for index in range(5):
    mu2 = -mu[index]
    mu3 = mu[index]
    ax1.plot(cauchy.cdf((x-mu3)/sigma3),cauchy.cdf((x-mu2)/sigma2))
    
plt.title('ROC Curves for Laplace Distribution')
plt.show()
```
