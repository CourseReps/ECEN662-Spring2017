```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.mlab as mlab
import math
from scipy.stats import laplace

fig, ax1 = plt.subplots()
mu0 = -1
variance0 = 1
sigma0 = math.sqrt(variance0)
x = np.linspace(-4, 4, 200)
ax1.plot(x,laplace.pdf(x, mu0, sigma0))

mu1 = 1
variance1 = 1
sigma1 = math.sqrt(variance1)
x = np.linspace(-4, 4, 200)
ax1.plot(x,laplace.pdf(x, mu1, sigma1),color='r')

ax2 = ax1.twinx()
ax2.plot(x,np.log(laplace.pdf(x, mu1, sigma1)) - np.log(laplace.pdf(x, mu0, sigma0)),'k--')

plt.show()
```

```python
fig2, ax1 = plt.subplots()

x = np.linspace(-25, 25, 200)
mu = np.linspace(0.25, 1.5, 6)
variance0 = 1.0
sigma0 = math.sqrt(variance0)
variance1 = 1.0
sigma1 = math.sqrt(variance1)
for index in range(5):
    mu0 = -mu[index]
    mu1 = mu[index]
    ax1.plot(laplace.cdf((x-mu1)/sigma1),laplace.cdf((x-mu0)/sigma0))

ax1.set_title('ROC for Laplace Distribution')    
ax1.set_xlabel('Probability of false positives')
ax1.set_ylabel('Probability of Detection')
plt.show()
```
