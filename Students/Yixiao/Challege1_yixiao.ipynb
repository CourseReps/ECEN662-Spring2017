# ROC Curves for Laplace and Cauchy Distributions

## pdf of Laplace with shifted mean = -1

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.mlab as mlab
import math

fig, ax1 = plt.subplots()
loc,scale = -1.,1.
l = np.random.laplace(loc, scale, 1000)
x = np.arange(-6., 6., .01)
l_pdf = np.exp(-abs(x-loc)/scale)/(2.*scale)
ax1.plot(x, l_pdf)
plt.show()
```

## pdf of Cauchy 

```python
from scipy.stats import cauchy
fig, ax = plt.subplots(1, 1)
x = np.linspace(cauchy.ppf(0.01),cauchy.ppf(0.99), 100)
ax.plot(x, cauchy.pdf(x), label='cauchy pdf')
plt.show()
```

## ROC Curves

```python
from scipy.stats import laplace

fig, ax1 = plt.subplots()
x1 = np.linspace(laplace.ppf(0.01), laplace.ppf(0.99), 100)
mu = np.linspace(0.25, 1.5, 6)
for index in range(5):
    mu0 = -mu[index]
    mu1 = mu[index]
    ax1.plot(laplace.cdf(x1,mu1, 1),laplace.cdf(x1,mu0,1))
plt.xlabel('Probability of False Alarm')
plt.ylabel('Probability of Detection')
plt.show()
```

```python
from scipy.stats import cauchy

fig, ax1 = plt.subplots()

x = np.linspace(cauchy.ppf(0.01), cauchy.ppf(0.99), 1000)
mu = np.linspace(0.25, 1.5, 6)

for index in range(5):
    mu0 = -mu[index]
    mu1 = mu[index]
    ax1.plot(cauchy.cdf(x,mu1,1),cauchy.cdf(x,mu0,1))
plt.xlabel('Probability of False Alarm')
plt.ylabel('Probability of Detection')
plt.show()
```
