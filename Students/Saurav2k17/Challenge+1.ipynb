## ROC curves for Laplace Distribution with shifted means

```python
#Receiver Operating Characteristic curves for Laplace Distribution with shifted means

from scipy.stats import laplace
import matplotlib.pyplot as plt
import numpy as np
import math

fig1,ax1=plt.subplots()

x = np.linspace(-4, 4, 200)
mu = np.linspace(0.25, 1.5, 6) 

for index in range(5):
    mu0 = -mu[index]
    mu1 = mu[index]
    ax1.plot(laplace.cdf(x,loc=mu1,scale=1),laplace.cdf(x,loc=mu0,scale=3))

plt.xlabel('False Positive Rate (1-Specificity) --->')
plt.ylabel('True Positive Rate (Sensitivity)--->')
plt.title('ROC curves for Laplace Distribution with shifted means')
plt.show()
```
