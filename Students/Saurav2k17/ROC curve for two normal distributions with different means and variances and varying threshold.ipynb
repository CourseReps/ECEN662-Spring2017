```python
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.mlab as mlab
import math

#First normal distribution with mean 2 and variance 20
fig1, ax1 = plt.subplots()
mu0 = 2
variance0 = 20
sigma0 = math.sqrt(variance0)
x = np.linspace(0, 30, 200)
ax1.plot(x,mlab.normpdf(x, mu0, sigma0)) #Blue curve

#Second normal distribution with mean 10 and variance 25
mu1 = 10
variance1 = 25
sigma1 = math.sqrt(variance1)
ax1.plot(x,mlab.normpdf(x, mu1, sigma1),color='g') #Green curve

ax2 = ax1.twinx()
ax2.plot(x,np.log(mlab.normpdf(x, mu1, sigma1)) - np.log(mlab.normpdf(x, mu0, sigma0)),'k--')
plt.title('PDF of 2 Normal Distributions with different means and variances and the log likelihood ratio')
plt.show() # Dotted curve is the log likelihood ratio


fig2,ax1 = plt.subplots()
t = np.linspace(4, 12, 6)
for index in range(6):
    threshold=np.linspace(t[index],30,200)
    ax1.plot(norm.cdf(threshold,mu1,sigma1),norm.cdf(threshold,mu0,sigma0))
    
plt.xlabel('False Positive Rate (1-Specificity) --->')
plt.ylabel('True Positive Rate (Sensitivity) --->')
plt.title('ROC curves for Normal Distribution with different means, variances and varying threshold')
plt.show()
```
